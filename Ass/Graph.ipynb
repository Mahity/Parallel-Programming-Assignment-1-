# Python Code For Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Problem ONE

```python
# Read the CSV files
part_a_df = pd.read_csv('part_a_output.csv')
slow_overhead_df = pd.read_csv('slow_overhead_output.csv')
fast_overhead_df = pd.read_csv('fast_overhead_output.csv')

# Plotting Part A Speedup and Efficiency
plt.figure(figsize=(12, 6))
for n in part_a_df['n'].unique():
    subset = part_a_df[part_a_df['n'] == n]
    plt.plot(subset['p'], subset['Speedup'], label=f'Speedup (n={n})')
plt.title('Speedup vs. Number of Processors (Part A)')
plt.xlabel('Number of Processors (p)')
plt.ylabel('Speedup')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
for n in part_a_df['n'].unique():
    subset = part_a_df[part_a_df['n'] == n]
    plt.plot(subset['p'], subset['Efficiency'], label=f'Efficiency (n={n})')
plt.title('Efficiency vs. Number of Processors (Part A)')
plt.xlabel('Number of Processors (p)')
plt.ylabel('Efficiency')
plt.legend()
plt.grid(True)
plt.show()

# Plotting Slow Overhead Growth
plt.figure(figsize=(12, 6))
for n in slow_overhead_df['n'].unique():
    subset = slow_overhead_df[slow_overhead_df['n'] == n]
    plt.plot(subset['p'], subset['Speedup'], label=f'Speedup (n={n})')
plt.title('Speedup vs. Number of Processors (Slow Overhead)')
plt.xlabel('Number of Processors (p)')
plt.ylabel('Speedup')
plt.legend()
plt.grid(True)
plt.show()

# Plotting Fast Overhead Growth
plt.figure(figsize=(12, 6))
for n in fast_overhead_df['n'].unique():
    subset = fast_overhead_df[fast_overhead_df['n'] == n]
    plt.plot(subset['p'], subset['Speedup'], label=f'Speedup (n={n})')
plt.title('Speedup vs. Number of Processors (Fast Overhead)')
plt.xlabel('Number of Processors (p)')
plt.ylabel('Speedup')
plt.legend()
plt.grid(True)
plt.show()
```

### Problem FOUR

```python

# Data
threads = [1, 2, 4, 8, 16]
pthreads_speedup = [0.997344, 1.980549, 3.094126, 3.928834, 3.713049]
openmp_speedup = [0.889346, 1.757215, 2.922038, 3.539373, 3.542681]
pthreads_efficiency = [0.997344, 0.990275, 0.773532, 0.491104, 0.232066]
openmp_efficiency = [0.889346, 0.878608, 0.730509, 0.442422, 0.221418]

# Speedup Plot
plt.figure()
plt.plot(threads, pthreads_speedup, marker='o', label='Pthreads')
plt.plot(threads, openmp_speedup, marker='o', label='OpenMP')
plt.xlabel("Number of Threads")
plt.ylabel("Speedup")
plt.title("Speedup vs. Number of Threads")
plt.legend()
plt.grid()
plt.show()

# Efficiency Plot
plt.figure()
plt.plot(threads, pthreads_efficiency, marker='o', label='Pthreads')
plt.plot(threads, openmp_efficiency, marker='o', label='OpenMP')
plt.xlabel("Number of Threads")
plt.ylabel("Efficiency")
plt.title("Efficiency vs. Number of Threads")
plt.legend()
plt.grid()
plt.show()
```
